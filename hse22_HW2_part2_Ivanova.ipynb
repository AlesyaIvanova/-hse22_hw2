````python
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqUtils import GC
from Bio.Data import CodonTable
from Bio import Entrez
from Bio.SeqFeature import SeqFeature, FeatureLocation
from pprint import pprint
````

````python
records = {}
for record in SeqIO.parse("scaffolds.fasta", "fasta"):
  record.annotations['molecule_type'] = 'DNA'
  records[record.id] = record
````

## Добавляем координаты генов и трансляции генов (аминокислотные посл-ти)

````python
genes = {}
for record in SeqIO.parse("proteins.fasta", "fasta"):
  locus, scaffold, start, end, strand = record.description.strip().split()[:5]
  gene_seq = record.seq
  if strand == '-':
    strand = -1
  else:
    strand = 1
  feature = SeqFeature(FeatureLocation(int(start), int(end), strand), type="CDS")
  feature.qualifiers['locus_tag'] = [locus]
  feature.qualifiers['translation'] = [gene_seq]
  records[scaffold].features.append(feature)
  genes[locus] = feature
````

````python
genes_mil_1 = {}
for feature in SeqIO.read("T_oleivorans_MIL_1.gbk", "genbank").features:
  if 'protein_id' not in feature.qualifiers or 'product' not in feature.qualifiers:
    continue
  genes_mil_1[feature.qualifiers['protein_id'][0]] = feature.qualifiers['product'][0]
````

````python
import pandas as pd
````

````python
columns = ['qseqid', 'sseqid', 'pident', 'length', 'mismatch', 'gapopen', 'qstart', 'qend', 'sstart', 'send', 'evalue', 'bitscore']
hits_mil_1 = pd.read_csv("scaffolds.hits_from_MIL_1.txt", sep='\t', header=None, names=columns)
hits_mil_1
````

````python
indexes = hits_mil_1['sseqid'].str.contains("CCU")
hits_mil_1 = hits_mil_1[indexes].sort_values('bitscore', ascending=False)
hits_mil_1.drop_duplicates('qseqid', inplace=True)
````

````python
for i, hit in hits_mil_1.iterrows():
  sequence = hit['sseqid'].split('_')[2]
  genes[str(hit['qseqid'])].qualifiers['product'] = [genes_mil_1[sequence]]
````

````python
for gen_loc in genes:
  if not genes[gen_loc].qualifiers.get('product', None):
    hits_mil_1 = hits_mil_1[hits_mil_1["qseqid"] != gen_loc]
````

````python
name = ""
genes_swiss = {}
for line in open('uniprot_sprot.dat'):
  if line[:2] == 'ID':
    name = line.split()[1]
  if line[:12] == 'DE   RecName':
    genes_swiss[name] = line.split('=')[1].split(';')[0]
````

````python
genes_swiss
````

````python
hits_swiss = pd.read_csv('scaffolds.hits_from_SwissProt.txt', sep='\t', header=None, names=columns)
hits_swiss = hits_swiss.sort_values('bitscore', ascending=False)
hits_swiss.drop_duplicates('qseqid', inplace=True)
````

````python
for i, hit in hits_swiss.iterrows():
  sequence = hit['sseqid'].split('|')[-1]
  genes[str(hit['qseqid'])].qualifiers['product'] = [genes_swiss[sequence]]
````
